# Findings
The data shows a disproportionate tendancy toward gun deaths in Males that are either White or Black. Minorities, Blacks aside, have relatively lower tendancy; with Hispanics being the next highest ranking at less than half the documented cases of blacks.

```python
intents = [x[3] for x in data]
races = [x[7] for x in data]
```

```python
homicide_race_counts = {}
for i,race in enumerate(races):
    if race not in homicide_race_counts:
        homicide_race_counts[race] = 1
    if intents[i] == "Homicide":
        homicide_race_counts[race] += 1
homicide_race_counts    
```

```python
race_per_hundredK = {}
for k,v in homicide_race_counts.items():
    race_per_hundredK[k] = (v / mapping[k]) * 100000
race_per_hundredK
```

# Findings
Findings show that nearly half of gun deaths in Blacks were considered to be homicide. Gun deaths with homicidal intent for all other races were quite low.

```python
months = [x[2] for x in data]
homicide_months_counts = {}
for i,month in enumerate(months):
    if month not in homicide_months_counts:
        homicide_months_counts[month] = 1
    if intents[i] == "Homicide":
        homicide_months_counts[month] += 1
homicide_months_counts  
```
