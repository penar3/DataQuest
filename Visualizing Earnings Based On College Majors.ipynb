# Histograms


```python
recent_grads.sort_values(by= ["Total"],ascending=False)
```

```python
SSmax=recent_grads['Sample_size'].max()
SSmin=recent_grads['Sample_size'].min()
ss=recent_grads["Sample_size"].hist(bins='doane',range=(SSmin,SSmax))
ss.set_xlabel('Sample Sizes')
ss.set_ylabel('Number of Samples')
ss.set_title('Sample Size')
```

```python
Medmax=recent_grads['Median'].max()
Medmin=recent_grads['Median'].min()
MedAx=recent_grads["Median"].hist(bins='sturges',range=(Medmin,Medmax))
MedAx.set_xlabel('Number of Emps')
MedAx.set_ylabel('Median Salary')
MedAx.set_title('Median Salary of FT Workers')
```

## Most Common Median Salary Range
The histogram above shows that the most common median salary range is roughly $30k to $40k with over 90 of the 172 major being in this range. 

```python
Empmax=recent_grads['Employed'].max()
Empmin=recent_grads['Employed'].min()
EmpAx=recent_grads["Employed"].hist(bins='sturges',range=(Empmin,Empmax))
EmpAx.set_xlabel('Number of Emps')
EmpAx.set_ylabel('Employed')
EmpAx.set_title('Number Employed')
```

```python
FTmax=recent_grads['Full_time'].max()
FTmin=recent_grads['Full_time'].min()
FtAx=recent_grads["Full_time"].hist(bins='sturges',range=(FTmin,FTmax))
FtAx.set_xlabel('Number of FT Emps')
FtAx.set_ylabel('Employed-FT')
FtAx.set_title('Number employed 35 hours or more')
```
